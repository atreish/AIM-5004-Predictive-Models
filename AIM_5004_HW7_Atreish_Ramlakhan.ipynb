# B

```python
#Define Log Likelihood function for one regressor B

def pi(x,B):
    return 1/(1+np.exp(-np.dot(x,B)))

def log_likelihood(x,y,B):
    return np.sum(((pi(x,B))**y)*((1-pi(x,B))**(1-y)))
```

# C

```python
# Generate some data
n = 100
x = np.random.uniform(size=(n,))
y = 1*(x+.1*np.random.normal(size=(n,))>.5)
```

# D

```python
B_test = log_likelihood(x,y,.3)
B_test
```

```python
#Test log_likelihood function on different values of B
Beta = np.arange(-10,10,0.0001)

ll_B = []
for B in Beta:
    lnLB = log_likelihood(x,y,B)
    ll_B.append(lnLB)
    
ll_B = np.asarray(ll_B)
print('The max probability for ML estimated regressor is {}'.format(ll_B.max()))
```

```python
plt.figure(figsize=(10,6))
plt.title('Log-Likelihood of Each Beta')
plt.xlabel('Beta')
plt.ylabel('Log Likelihood')
plt.axhline(y=ll_B.max(), color='r', linestyle='dashed')
plt.plot(Beta,ll_B)
plt.show()
```
