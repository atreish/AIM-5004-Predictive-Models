```python
#Array of Mu candidates 
mu_candidates = np.arange(0, 2.01, .1)

#Generate our sample vector where the Mean = 1
sample_vector = [1,-4,2,5,-2,-1,3,1,4,1]

#Ensure that the mean=1
mu = sum(sample_vector) / len(sample_vector)
mu
```

```python
'''Define a function where we can create an Error vector of the Sum of Squared Errors for a given mean, which we
will eventually take from the Mu vector, then can analyze the results'''
def SSE(mean):
    list_error = [(i-mean)**2 for i in sample_vector]
    print('\nError Vector for Mu={}: \n'.format(mean), list_error)
    SSE = sum(list_error)
    return SSE
```

```python
#apply our function to all the 
SSE_for_all_Mu = [SSE(i) for i in mu_candidates]
print('\nSSE_vector:', SSE_for_all_Mu)
```

```python
#Create a plot that shows the values of the Sum of Squared Errore against values of Mu
plt.figure(figsize=(12, 6))
plt.xlabel("Values of Mu vector") 
plt.ylabel("Sum of Squared Errors")
plt.xticks(mu_candidates)
plt.plot(mu_candidates, SSE_for_all_Mu)
plt.show()
```

Observe, that the SAMPLE MEAN, of Mu=1, does in fact minimize the Sum of Squared Errors that is Hypothesized
